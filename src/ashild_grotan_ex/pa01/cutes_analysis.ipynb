```python
from snakes_and_ladders import multi_game_experiment
import matplotlib.pyplot as plt
```

```python
games_1 = multi_game_experiment(100, 1, 1)
games_2 = multi_game_experiment(100, 2, 1)
games_3 = multi_game_experiment(100, 4, 1)
games_4 = multi_game_experiment(100, 8, 1)
```

```python
plt.hist(games_1, bins='auto', color='#0504aa', histtype='step')
plt.xlabel('Turns')
plt.ylabel('Frequency')
plt.title('Turns per game with 1 player')
plt.show()
```

```python
plt.hist(games_2, bins='auto', color='#0504aa', histtype='step')
plt.xlabel('Turns')
plt.ylabel('Frequency')
plt.title('Turns per game with 2 players')
plt.show()
```

```python
plt.hist(games_3, bins='auto', color='#0504aa', histtype='step')
plt.xlabel('Turns')
plt.ylabel('Frequency')
plt.title('Turns per game with 4 players')
plt.show()

plt.hist(games_4, bins='auto', color='#0504aa', histtype='step')
plt.xlabel('Turns')
plt.ylabel('Frequency')
plt.title('Turns per game with 8 players')
plt.show()
```
